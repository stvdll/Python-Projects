```python
import feedparser
from transformers import pipeline
import pandas as pd

model = pipeline("text-classification", model="ProsusAI/finbert") ##Fine tuned NLP Modell which is trained on lots of financial text.

url = f'https://finance.yahoo.com//rss/headlne?s'

feed = feedparser.parse(url)


data = []
dates = []

for i,entry in enumerate(feed.entries):

  if i<100:

    sentiment = model(entry.title)[0]

    sentiment['dates'] = entry.published

    data.append(sentiment)


data = pd.DataFrame(data)

n, neg, po = data['label'].value_counts()

neutralscores = data.loc[data['label'] == 'neutral', 'score'].mean()
positivscores = data.loc[data['label'] == 'positive', 'score'].mean()
negativscores = data.loc[data['label'] == 'negative', 'score'].mean()




print(f'Postiv {po/(len(data))} mit {positivscores} Sicherheit')
print(f'Negativ {neg/(len(data))} mit {negativscores} Sicherheit')
print(f'neutral {n/(len(data))} mit {neutralscores} Sicherheit')

print('Gesamteinträge', (len(data)))
```
